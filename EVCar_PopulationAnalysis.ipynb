```python
# Check which columns have null values
null_counts = df.isnull().sum()
print(null_counts[null_counts > 0])
```

```python
# Fill missing values and assign back to the DataFrame
df['County'] = df['County'].fillna('Unknown')
df['City'] = df['City'].fillna('Unknown')
df['Electric Utility'] = df['Electric Utility'].fillna('Unknown')
df['Vehicle Location'] = df['Vehicle Location'].fillna('Unknown')
```

```python
# Fill missing values with a specific numeric code
df['Postal Code'] = df['Postal Code'].fillna(99999)
df['Legislative District'] = df['Legislative District'].fillna(-1)
df['2020 Census Tract'] = df['2020 Census Tract'].fillna(9999999999)
```
